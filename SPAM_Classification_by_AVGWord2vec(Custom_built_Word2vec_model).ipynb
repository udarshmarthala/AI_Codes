```python
import pandas as pd
import regex as re
messages =pd.read_csv('SMSSpamCollection.txt', sep='\t',names=["label", "message" ])
```

```python
messages
```

```python
messages.shape
```

```python
import nltk
from nltk.stem import PorterStemmer#for stemming
from nltk.corpus import stopwords
stemm=PorterStemmer()
```

```python
new_sentences=[]
for i in range(len(messages["message"])):
    temp=re.sub('[^a-zA-z]',' ',messages["message"][i])
    temp=temp.lower()
    temp=temp.split(" ")
    temp=[stemm.stem(i) for i in temp if i not in stopwords.words("english")]
    temp=" ".join(temp)
    new_sentences.append(temp)
```

```python
new_sentences
```

```python
if("until" in stopwords.words("english")):
    print(2)
```

# just checking the accuracy from using some other model like naivebayes using TF-IDF

```python

from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer (max_features=3000)
x = tv.fit_transform(new_sentences).toarray()
```

```python
x
```

```python
y=pd.get_dummies(messages['label'],dtype=int)
```

```python
y
```

```python
y=y["spam"]
```

```python
y
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
